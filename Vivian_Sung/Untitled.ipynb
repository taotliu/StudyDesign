```r
library(pwr)
```

```r
eff.size = seq(14, 26, 1)

foo = pwr.t.test(n=100, sig.level = 0.025, eff.size/30)
foo2 = pwr.t.test(n=100, sig.level = 0.025, eff.size/25)
```

```r
plot(eff.size, foo$power*100, type = "b", ylim=c(50, 100), 
    ylab="Power (%)", 
    xlab="Effect Size", 
    main="OABq-SS score")
#lines(eff.size, foo2$power*100, type = "b", col = "red")

#legend(22, 60, 
#       legend = c("SD = 30", "SD = 25"), bty="n", 
#       lty=c(1,1), col=c("black", "red"), pch=c(1,1)
#      )

legend(21.4, 67, 
       legend = c("N = 100 / group", 
                  "Sig Level = 0.025"), bty="n"
      )
```

```r
p2 = seq(0.40, 0.55, 0.01)

h = 2*( asin(sqrt(0.64)) - asin(sqrt(p2)) )
foo = pwr.2p.test(n=90, sig.level = 0.025, h = h)
```

```r
plot((0.75-p2)*100, foo$power*100, type = "b", ylim=c(50, 100), 
    ylab="Power (%)", 
    xlab="Effect Size (%)")
abline(h=80, lty=3)
legend(26, 63, 
       legend = c("N = 90 / group", 
                  "Sig Level = 0.025"), bty="n"
      )
```
